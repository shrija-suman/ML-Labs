```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")

print(df.head())
```

```python
print(df.isnull())
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
print("Original shape:", df.shape)

df_clean = df.dropna()

print("After removing missing rows:", df_clean.shape)
```

```python
df_fill = df.copy()

df_fill['total_bill'] = df_fill['total_bill'].fillna(df_fill['total_bill'].mean())
df_fill['sex'] = df_fill['sex'].fillna("Unknown")

print(df_fill.isnull().sum())
```

```python
print(df.dtypes)

df['size'] = df['size'].astype(float)

print("\nAfter conversion:")
print(df.dtypes)
```

```python
df_renamed = df.rename(columns={"total_bill":"TotalBill","tip":"TipAmount"})

print(df_renamed.head())
```

```python
df_replace = df.copy()

df_replace['sex'] = df_replace['sex'].replace({'Male':'M','Female':'F'})

print(df_replace.head())
```

```python
plt.figure(figsize=(6,4))
plt.plot(df['total_bill'])
plt.title("Line Plot of Total Bill")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=df)
plt.title("Average Bill per Day")
plt.show()
```

```python
plt.scatter(df['total_bill'], df['tip'])

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.title("Total Bill vs Tip")

plt.show()
```

```python
plt.hist(df['total_bill'], bins=20)

plt.title("Distribution of Total Bill")

plt.show()
```

```python
sns.boxplot(x=df['total_bill'])

plt.title("Boxplot of Total Bill")

plt.show()
```
