```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"

df = pd.read_csv(url)

print(df.head())
```

```python
print(df.info())
```

```python
numeric_cols = ["math score","reading score","writing score"]

print(df[numeric_cols].head())
```

```python
print("Mean:")
print(df[numeric_cols].mean())

print("\nMedian:")
print(df[numeric_cols].median())

print("\nMode:")
print(df[numeric_cols].mode().iloc[0])
```

```python
print("Minimum:")
print(df[numeric_cols].min())

print("Maximum:")
print(df[numeric_cols].max())

print("Sum:")
print(df[numeric_cols].sum())

print("Variance:")
print(df[numeric_cols].var())

print("Standard Deviation:")
print(df[numeric_cols].std())
```

```python
Q1 = df[numeric_cols].quantile(0.25)
Q2 = df[numeric_cols].quantile(0.50)
Q3 = df[numeric_cols].quantile(0.75)

print("First Quartile (Q1):")
print(Q1)

print("\nSecond Quartile (Median Q2):")
print(Q2)

print("\nThird Quartile (Q3):")
print(Q3)
```

```python
corr_matrix = df[numeric_cols].corr()

print(corr_matrix)
```

```python
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
cov_matrix = df[numeric_cols].cov()

print("Covariance Matrix:")
print(cov_matrix)
```

```python
df[numeric_cols].hist(figsize=(8,5))

plt.suptitle("Histogram of Scores")

plt.show()
```

```python
sns.boxplot(data=df[numeric_cols])

plt.title("Boxplot of Student Scores")

plt.show()
```
